Suppose $x \sim \rho()$ and y = f(x) is a determinestic transformation of it.

This section discusses how to compute $\rho(y)$, the new transformed density distribution

# Discrete transformation

We can just sum the probability mass for all X's such that f(x)=y

$\rho_y(y) = \sum_{x:(f(x)=y)} \rho_x(x)$


## Example

Let $X\sim \rho_x(x)$ 

Note, p is sub (x) because its the pdf for the dimension of x, and it takes in x as an input

Our transformation of X, takes X to a new distribution, Y

let $Y = aX+b$

Such that $Y\sim\rho_y(y)$

We want to find $\rho_y(y)$, which is the pmf after our linear transformation aX+b

### Proof

Y is a discrete random variable that is distributed across $\real$

$Pr(Y=y)$, means, "The probability that Y, our discrete random variable, is equal to a specific y in $\real$"

The probability that Y is a given random variable $y$ that is is allowed to be, is equivilant to the input of $y$ in $\rho_y(y)$

Therefore,

$\rho_y(y) = Pr(Y=y)$, The pdf value of the value y is equivilant to the probability that a given value $Y$ is equal to $y$

We know what Y is by mapping it through f(x)=Y

$Pr(Y=y) = Pr(aX+b=y)$, what this means that the probability that our discrete value Y is equal to any value y, is equal to its function mapping

We can re-arrange this to give us the Probability in terms of X

$Pr(X = \frac{y-b}{a})$. Now we can use the same definitional trick as before. If $X\sim\rho_x(x)$, then the probability of observing a given value X is given by $\rho_x(x)$ 

(using 
$\rho_y(y) = Pr(Y=y)$, as our guide....)

Therefore...

$Pr(X = \frac{y-b}{a})$ -> $\rho_x (\frac{y-b}{a})$

and...

$\rho_y(y) = \rho_x(\frac{y-b}{a})$

### Example 1

Suppose we want to transform the following discrete distribution

$ X \in \{-2...2\}$

$ \rho_x(x) = 1/5 $, for -2,-1,0,1,2

Let Y = 2X+2

$ Y \in 2*\{-2,-1,0,1,2\} + 2 -> \{-2,0,2,4,6\} $

$\rho_y(y) = \rho_x(\frac{y-2}{2})$

$\rho(y) = \rho_x(-2,-1,0,1,2) -> [1/5, 1/5, 1/5, 1/5, 1/5]$

So... The transformation is now in $Y~\in\{-2,0,2,4,6\}$ and the distribution remained uniform

### Note. 

a (slope) of the linear transformation broadens the discrete distribution

b of the linear transformaiton transforms it in space.

### Example 2
Lets do another linear transformation 
Let $ X \in \{-2,0,2,4,6\}$ and $\rho_x(x) = 1/5$

Let $Y= X^2$, find $\rho_y(y)$

1. Find Y~f(x)...

$ Y \in \{-2,0,2,4,6\}^2 ->\{4,0,4,16,36\} ->\{0,4,16,36\}$

2. Use transformation formula,

$ \rho_y(y) = Pr(Y=y) = Pr(X^2=y) = Pr(X = \sqrt y) = \rho_x(\sqrt y)$

Therefore,

$\rho_y(y=36) = \rho_x(\sqrt 36) = \rho_x(+/-6), \rho_x(6)=1/5$
$\rho_y(y=16) = \rho_x(\sqrt 16) = \rho_x(+/-4), \rho_x(4)=1/5$
$\rho_y(y=4) = \rho_x(\sqrt 4) = \rho_x(+/-2), \rho_x(2)=1/5, \rho_x(-2)=1/5 -> 2/5 $
$\rho_y(y=0) = \rho_x(\sqrt 0) = \rho_x(0), \rho_x(0)=1/5, 

So

$\rho_y(y) = [1/5, 2/5, 1/5, 1/5]$
for $ Y \in \{0,4,16,36\}$


# Linear Transformations with Continuous variables

If X is continuous we cannot jst sum.

We have to use the cdf

### Proof

Suppose X is continus random variable with pdf $\rho_x(x)$

Let Y = aX+b

Then...

$\rho_y(y) = \frac{1}{|a|} \rho_x(\frac{y-b}{a})$

via the following proof

Y 

$P_y(y) = Pr(Y<=y)$. By definition, the value for y shoved into the cdf gives you the probability of observing a value less than or equal to y.

Note that Y here is the discrete space of posible y values. So its like saying that the cdf(y) gives the probability that any value of Y is less than or equal to y.

Consequently, because Y represents the space here (which is defined by aX+b)...

$Pr(Y<=y) = Pr(aX+b<=y)$. We know $\rho_x(x)$, so lets re-arange the formula...

-> $Pr(aX < = y-b)$. We need to think about wether a is + or negative becuase its its negative the direction of the inequality swaps...

### Case a>0

$P_y(y) = Pr(X <= \frac{y-b}{a})$ 

Via the same definitional formula of what it means to be a cdf...

$Pr(X <= \frac{y-b}{a})= P_x (\frac{y-b}{a})$

Recall, pdf = derivative of cdf

$\frac{d}{d_y} P_y(y) = \rho_y(y)$

$\rho_y(y) = \frac{d}{d_y} P_x (\frac{y-b}{a})$ = $\frac{1}{a} \rho_x (\frac{y-b}{a})$

Note, this last step is confusing because of the x subscript and you are taking a y derivative, but abstractly we actually have $\frac{d}{d_y} f(y)$










# Change of Variables Formula

Using the change of variables formula, we can transform continus variable's pdf for 1 location (infitestesamally small) in another space

Suppose $\chi \sim Unif(0,1)$ and $y=2x+1$.

Lets zoom in on a point and another infinitesimally close point named $x+d_x$.

We see that this would get mapped to an infetesimally small segment in y.

1. The probability mass of these intervals must be the esame

$\rho_x(x)d_x = \rho_y(y)d_y$

Therefore

$\rho_y(y) = \rho_x(x)\frac{d_x}{d_y}$, here we can actually move $d_y$ because it representa a small change in y

Now, consider the monotonif function $f$.

Let $g = f^{-1}x$ (inversion of f).

Therefore if $y=f(x)$, then by difinition, $x = f^{-1}(y) -> x = g(y)$

Therefore, 

$P_y(y)$ = P_x(g(y))$

After some derivations with respect to y

$\rho_y(y) = \rho_x(g(y)) | \frac{d}{d_y}(g(y))|$

using this change of variables formula we can transform any pdf to another pdf for monotonic and hance invertible functions.

# The Convolution Theorum

Let $ y = x_y + x_2$ where $x_1$ and $x_2$ are discrete independent rvs with known pmfs

We can compute the pmf for the sum of x1 and x2.

This is called their convolution

### pmf for added rvs that are discrete

$\rho(y = j) = \sum_k \rho(x_1=k) \rho(x_2 = j-k)$

for $j \in${sum of rvs}

### pmf for continuous

We again have to go through the cdf.

We show that the value of y given by the cdf is equal to the integral of the pdf

$\rho_y(y) = \int \rho_1(x_1)\rho_2(y-x_1)dx_1$.

Here we integrate over all possible values of x11. Each contribution is weighted by the two pmfs


## THe convolution operator

we can re-write the previous equation as follows

$\rho(y) = \rho_1 ⊛ \rho_2$

### Example with discrete uniform distribution

# Monte Carlo Approximation

Suppose x is a random variable, and y=f(x).

It is often really hard to analytically compute the induced distribution $\rho(y)$

One way this is often done instead is to just get a large numberof samples (x), push threm through y to approximate $\rho(y)$

This is called a Monte Carlo approximation
