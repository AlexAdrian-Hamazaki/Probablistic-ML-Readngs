# Bernoulli Distribution

Used to model the probability of binary events occuring

Coin toss
Let Y =1 denote heads
Y=0 denote tails

$\rho(Y)$ denotes the probability distribution of probability the coin lands on heads or tails

Let $\theta$ = the probability of heads = $\rho(Y=1)$

This means that $\rho(Y=0)$ = $1-\theta$

These two things define the Bernouli

Y ~ $Ber(\theta)$, notation wise means that Y follows a bernouli distribution

# Binomial Distribution

An expansion of the bernouli distribution...

suppose we observe N bernouli trials (tossing N coins).

Let s be the number of observed heads in N trials

$\mathbb{1}(Y=1)$ : Means "1" if the following condition is met

$s = \sum_{n=1}^N \mathbb{1}(y_n=1)$, count the number of heads

There is randomness in how many heads we are going to get in N counts,

The distribution of s is given by the binomial distribution:

$Bin(s|N, \theta) = \binom{N}{s}\theta^s(1-\theta)^{N-s}$

Where theta is the probability of s occuring

### Binomial Coefficient

Can be interpreted as the number of ways to choose k items from N.

If we have 10 flips, and s=2. it represents how many different ordered combinations we could sum to 2 heads

$\binom{N}{s} = \frac{N!}{(N-k)!k!}$

### Relationship between Binom and Bernouli

If N = 1, then the binom reduces to the bernouli

# Sigmoid/Logistic Function

When we want to predict the probability of a binary variable $y \in {0,1}$, for some inputs $x \in \chi$, we can do this by modeling a conditional probability distribution

$\rho(y|x, \theta) = Ber(y|f(x;\theta))$

Where $f(x;\theta)$ is some function that predicts the mean parameter of the bernouli distribution

### Constraining with the Sigmoid/Logistif function

$\sigma(a) = \frac{1}{1+e^{-a}}$

Adding this to our previous probability prediction for our y, we can constrain the output of $f(x:\theta)$ to be between 0-1.

Enabbling a proper probability modeling over y

$\rho(y|x, \theta) = Ber(y|\sigma (f(x;\theta)))$

where $a = f(x;\theta)$

# Binary Logistic Regression

$f(x;\theta) = \omega^Tx+b$

If so, we can model the probability of y given x using a bernouli

$\rho(y|x;\theta) = Ber(y|\sigma(\omega^Tx+b))$

This is the logistic regression formula...

$\rho(y=1|x;\theta) = Ber(1|\sigma(\omega^Tx+b)) = \frac{1}{1+e^{-(\omega^Tx+b)}}$
